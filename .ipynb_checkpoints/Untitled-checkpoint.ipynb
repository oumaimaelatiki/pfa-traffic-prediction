```python
import pandas as pd

# Chargement du fichier Excel
file_path = "Dataset for traffic analysis in Casablanca, Morocco.xlsx"
excel_file = pd.ExcelFile(file_path)

print("✅ Fichier chargé avec succès !")
print("Feuilles disponibles :", excel_file.sheet_names)
```

```python
# Chargement des tables statiques (données fixes par commune)
print("Chargement des tables statiques...")

# Table 0 : Coordonnées géographiques
df_coords_raw = pd.read_excel(file_path, sheet_name='Table 0. Coordinates', header=None)

# Table 1 : Population  
df_pop_raw = pd.read_excel(file_path, sheet_name='Table 1. Population size in eac', header=None)

# Table 2 : Transport (Tram/Bus)
df_transport_raw = pd.read_excel(file_path, sheet_name='Table 2. Number of Tram and Bus', header=None)

# Table 3 : Types de routes
df_roads_raw = pd.read_excel(file_path, sheet_name='Table 3. Type of roads', header=None)

# Table 4 : Usage des sols
df_landuse_raw = pd.read_excel(file_path, sheet_name='Table 4. Land use variables ', header=None)

print("✅ Tables statiques chargées :")
print(f"  - Coordonnées : {df_coords_raw.shape}")
print(f"  - Population : {df_pop_raw.shape}")
print(f"  - Transport : {df_transport_raw.shape}")
print(f"  - Routes : {df_roads_raw.shape}")
print(f"  - Usage sols : {df_landuse_raw.shape}")
```

```python
print("Nettoyage des coordonnées - Correction finale...")

# Recharger depuis le fichier
df_coords_raw = pd.read_excel(file_path, sheet_name='Table 0. Coordinates', header=None)

# Les headers sont à l'index 11, les données commencent à l'index 12
df_coords = df_coords_raw.iloc[12:].copy()
df_coords.columns = df_coords_raw.iloc[11]

# Supprimer la première colonne qui est vide (NaN)
if df_coords.columns[0] != df_coords.columns[0]:  # Vérifie si c'est NaN
    df_coords = df_coords.iloc[:, 1:].copy()

# Renommer les colonnes proprement
df_coords.columns = ['Commune', 'ZIP_code', 'Point_Index', 'Latitude', 'Longitude', 'Coordinates']

# Remplir les Communes et ZIP (forward fill)
df_coords['Commune'] = df_coords['Commune'].ffill()
df_coords['ZIP_code'] = df_coords['ZIP_code'].ffill()

# Convertir en numérique
df_coords['Latitude'] = pd.to_numeric(df_coords['Latitude'], errors='coerce')
df_coords['Longitude'] = pd.to_numeric(df_coords['Longitude'], errors='coerce')
df_coords['Point_Index'] = pd.to_numeric(df_coords['Point_Index'], errors='coerce')
df_coords['ZIP_code'] = pd.to_numeric(df_coords['ZIP_code'], errors='coerce')

# Supprimer les lignes sans coordonnées
df_coords = df_coords.dropna(subset=['Latitude', 'Longitude'])

print(f"✅ Coordonnées nettoyées : {len(df_coords)} points")
print(f"Nombre de communes : {df_coords['Commune'].nunique()}")
print(f"Communes : {sorted(df_coords['Commune'].unique())}")
print(f"\nAperçu :")
print(df_coords[['Commune', 'ZIP_code', 'Point_Index', 'Latitude', 'Longitude']].head(10))
```

```python
print("Nettoyage des tables statiques (corrigé)...")

# Recharger les données brutes
df_pop_raw = pd.read_excel(file_path, sheet_name='Table 1. Population size in eac', header=None)
df_transport_raw = pd.read_excel(file_path, sheet_name='Table 2. Number of Tram and Bus', header=None)
df_roads_raw = pd.read_excel(file_path, sheet_name='Table 3. Type of roads', header=None)

# Pour Population : Header à l'index 9, données à partir de l'index 10
# On ignore la première colonne (index 0) qui est vide
df_pop = df_pop_raw.iloc[10:, 1:5].copy()  # Lignes 10+, colonnes 1 à 4
df_pop.columns = ['Commune', 'ZIP_code', 'Population', 'Households']
df_pop = df_pop.dropna(subset=['Commune'])  # Supprimer les lignes sans commune
df_pop['ZIP_code'] = pd.to_numeric(df_pop['ZIP_code'], errors='coerce')
df_pop['Population'] = pd.to_numeric(df_pop['Population'], errors='coerce')
df_pop['Households'] = pd.to_numeric(df_pop['Households'], errors='coerce')

# Pour Transport : même logique
df_transport = df_transport_raw.iloc[10:, 1:4].copy()  # 3 colonnes utiles
df_transport.columns = ['Commune', 'ZIP_code', 'Tram_Stations']
df_transport = df_transport.dropna(subset=['Commune'])
df_transport['ZIP_code'] = pd.to_numeric(df_transport['ZIP_code'], errors='coerce')
df_transport['Tram_Stations'] = pd.to_numeric(df_transport['Tram_Stations'], errors='coerce').fillna(0)

# Pour Routes : même logique  
df_roads = df_roads_raw.iloc[10:, 1:5].copy()  # 4 colonnes utiles
df_roads.columns = ['Commune', 'ZIP_code', 'Primary_Roads', 'Secondary_Roads']
df_roads = df_roads.dropna(subset=['Commune'])
df_roads['ZIP_code'] = pd.to_numeric(df_roads['ZIP_code'], errors='coerce')
df_roads['Primary_Roads'] = pd.to_numeric(df_roads['Primary_Roads'], errors='coerce')
df_roads['Secondary_Roads'] = pd.to_numeric(df_roads['Secondary_Roads'], errors='coerce')

print(f"✅ Population : {len(df_pop)} communes")
print(f"✅ Transport : {len(df_transport)} communes")  
print(f"✅ Routes : {len(df_roads)} communes")

print("\nAperçu Population :")
print(df_pop.head())
print("\nAperçu Transport :")
print(df_transport.head())
```

```python
print("Fusion des données statiques...")

# Fusion progressive sur Commune et ZIP_code
df_static = df_pop.merge(df_transport, on=['Commune', 'ZIP_code'], how='outer')
df_static = df_static.merge(df_roads, on=['Commune', 'ZIP_code'], how='outer')

# Ajouter les coordonnées (nombre de points par commune)
coords_summary = df_coords.groupby('Commune').agg({
    'Point_Index': 'count',
    'Latitude': 'mean',
    'Longitude': 'mean'
}).reset_index()
coords_summary.columns = ['Commune', 'Num_Points', 'Avg_Latitude', 'Avg_Longitude']

df_static = df_static.merge(coords_summary, on='Commune', how='left')

print(f"✅ Dataset statique final : {df_static.shape}")
print(f"Colonnes : {list(df_static.columns)}")
print("\nAperçu :")
print(df_static.head())
```

```python
print("Chargement final des données de trafic...")

days_sheets = {
    'Monday': ('Table 5. Monday', 9),
    'Tuesday': ('Table 6. Tuesday', 11), 
    'Wednesday': ('Table 7. Wednesday', 11),
    'Thursday': ('Table. 8 Thursday', 11),
    'Friday': ('Table. 9 Friday', 11),
    'Saturday': ('Table. 10 Saturday', 11),
    'Sunday': ('Table. 11 Sunday', 11)
}

traffic_data = []

for day, (sheet_name, header_row) in days_sheets.items():
    print(f"\n  → {day}...")
    
    # Charger avec le bon header
    df_day = pd.read_excel(file_path, sheet_name=sheet_name, header=header_row)
    
    # Nettoyer noms de colonnes
    df_day.columns = [str(c).strip() for c in df_day.columns]
    
    # Renommer Commune et ZIP
    for col in list(df_day.columns):
        if 'COMMUNE' in col.upper():
            df_day = df_day.rename(columns={col: 'Commune'})
        elif 'ZIP' in col.upper():
            df_day = df_day.rename(columns={col: 'ZIP_code'})
    
    # Trouver les colonnes horaires (0-23 ou H0-H23)
    hour_cols = []
    rename_dict = {}
    
    for col in df_day.columns:
        col_str = str(col)
        # Format: "0", "1", ..., "23"
        if col_str.isdigit() and 0 <= int(col_str) <= 23:
            hour_cols.append(col)
            rename_dict[col] = f'H{col_str}'
        # Format: "H0", "H1"...
        elif col_str.startswith('H') and col_str[1:].isdigit():
            hour_cols.append(col)
    
    # Renommer les colonures numériques en H0, H1...
    if rename_dict:
        df_day = df_day.rename(columns=rename_dict)
    
    print(f"     {len(hour_cols)} heures trouvées | {len(df_day.columns)} colonnes total")
    
    # Nettoyer données
    df_day = df_day.dropna(subset=['Commune'])
    df_day = df_day[~df_day['Commune'].astype(str).str.contains('Unnamed|nan', case=False, na=False)]
    
    # Convertir heures en numérique
    for h in [f'H{i}' for i in range(24)]:
        if h in df_day.columns:
            df_day[h] = pd.to_numeric(df_day[h], errors='coerce')
    
    if 'ZIP_code' in df_day.columns:
        df_day['ZIP_code'] = pd.to_numeric(df_day['ZIP_code'], errors='coerce')
    
    df_day['Day'] = day
    traffic_data.append(df_day)
    print(f"     ✓ {len(df_day)} trajectoires")

# Concaténer
df_traffic = pd.concat(traffic_data, ignore_index=True)
print(f"\n✅ SUCCÈS: {len(df_traffic)} lignes sur {len(traffic_data)} jours")

# Vérifier les heures
hour_cols_check = [c for c in df_traffic.columns if str(c).startswith('H')]
print(f"Colonnes horaires: {len(hour_cols_check)} (de {hour_cols_check[0]} à {hour_cols_check[-1]})")

# Aperçu
if hour_cols_check:
    print(f"\n📊 Exemple de données H0:")
    print(df_traffic[['Commune', 'Day', 'H0', 'H12']].head())
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Préparer les données
hour_cols = [f'H{i}' for i in range(24)]
traffic_by_hour = df_final.groupby('Day')[hour_cols].mean()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
traffic_by_hour = traffic_by_hour.reindex([d for d in day_order if d in traffic_by_hour.index])

plt.figure(figsize=(16, 7))
ax = sns.heatmap(traffic_by_hour, cmap='YlOrRd', annot=False, fmt='.1f', 
                 linewidths=0.5, cbar_kws={'label': 'Distance moyenne (km)'})

plt.title('🔥 Heatmap des distances de trajet par heure et jour\n(Casablanca - 22 communes)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Heure de la journée', fontsize=12)
plt.ylabel('Jour de la semaine', fontsize=12)

# Améliorer l'affichage des heures
ax.set_xticks(np.arange(0, 24, 2))
ax.set_xticklabels([f'{i}h' for i in range(0, 24, 2)])
plt.xticks(rotation=0)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()

print("💡 Interprétation:")
print("- Rouge foncé = longues distances (trajets domicile-travail)")
print("- Jaune = courtes distances (déplacements locaux)")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Classement des communes par distance moyenne
commune_stats = df_final.groupby('Commune').agg({
    'Avg_Distance_km': 'mean',
    'Population': 'first',
    'Tram_Stations': 'first'
}).sort_values('Avg_Distance_km', ascending=True)

# Top 12 (plus éloignées)
commune_stats.tail(12)['Avg_Distance_km'].plot(kind='barh', ax=axes[0,0], color='steelblue')
axes[0,0].set_title('🏙️ Communes les plus éloignées (Top 12)', fontweight='bold', fontsize=12)
axes[0,0].set_xlabel('Distance moyenne (km)')

# 2. Impact du tram sur les distances
tram_impact = df_final.groupby('Tram_Stations')['Avg_Distance_km'].mean()
colors = ['red' if x == 0 else 'green' for x in tram_impact.index]
tram_impact.plot(kind='bar', ax=axes[0,1], color=colors, alpha=0.7)
axes[0,1].set_title('🚊 Impact du tram (0 = pas de tram)', fontweight='bold', fontsize=12)
axes[0,1].set_xlabel('Nombre de stations de tram')
axes[0,1].set_ylabel('Distance moyenne (km)')
axes[0,1].axhline(y=tram_impact.mean(), color='black', linestyle='--', label='Moyenne')
axes[0,1].legend()

# 3. Corrélation Population vs Distance (avec labels)
scatter = axes[1,0].scatter(df_final['Population'], df_final['Avg_Distance_km'], 
                           c=df_final['Tram_Stations'], s=100, alpha=0.6, cmap='viridis',
                           edgecolors='black', linewidth=0.5)
axes[1,0].set_xlabel('Population', fontsize=11)
axes[1,0].set_ylabel('Distance moyenne (km)', fontsize=11)
axes[1,0].set_title('👥 Population vs Distance (couleur = nb stations tram)', fontweight='bold', fontsize=12)
axes[1,0].grid(alpha=0.3)
plt.colorbar(scatter, ax=axes[1,0], label='Stations tram')

# Annoter les extrêmes
for idx, row in df_final.iterrows():
    if row['Avg_Distance_km'] > 7 or row['Population'] > 180000:
        axes[1,0].annotate(row['Commune'], 
                          (row['Population'], row['Avg_Distance_km']),
                          xytext=(5, 5), textcoords='offset points', fontsize=8, alpha=0.8)

# 4. Comparaison Weekday vs Weekend
df_final['Type_Jour'] = df_final['Day'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Semaine')
comparison = df_final.groupby(['Commune', 'Type_Jour'])['Avg_Distance_km'].mean().unstack()

# Prendre les 10 communes les plus peuplées pour la lisibilité
top_communes = df_final.groupby('Commune')['Population'].first().sort_values(ascending=False).head(10).index
comparison.loc[top_communes].plot(kind='bar', ax=axes[1,1], color=['skyblue', 'orange'])
axes[1,1].set_title('📅 Semaine vs Weekend (Top 10 communes)', fontweight='bold', fontsize=12)
axes[1,1].set_ylabel('Distance moyenne (km)')
axes[1,1].legend(title='Type de jour')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Statistiques détaillées
print("📈 STATISTIQUES CLÉS:")
print(f"1. Commune la plus éloignée: {commune_stats.index[-1]} ({commune_stats.iloc[-1]['Avg_Distance_km']:.2f} km)")
print(f"2. Commune la plus centrale: {commune_stats.index[0]} ({commune_stats.iloc[0]['Avg_Distance_km']:.2f} km)")
print(f"3. Impact du tram: Avec tram = {df_final[df_final['Tram_Stations']>0]['Avg_Distance_km'].mean():.2f} km")
print(f"   Sans tram = {df_final[df_final['Tram_Stations']==0]['Avg_Distance_km'].mean():.2f} km")
print(f"4. Différence Semaine/Weekend: {comparison['Semaine'].mean() - comparison['Weekend'].mean():.2f} km")
```

```python
import pandas as pd

print("📊 CRÉATION DU FICHIER EXCEL MASTER")
print("=" * 50)

# Créer un fichier Excel avec plusieurs feuilles
with pd.ExcelWriter('CASABLANCA_DATA_NETTOYEES.xlsx', engine='openpyxl') as writer:
    
    # FEUILLE 1 : Données complètes (comme ton image mais enrichi)
    print("→ Feuille 1: DONNÉES COMPLÈTES...")
    
    # Préparer les données comme ton image + trafic
    df_master = df_final.groupby('Commune').agg({
        'ZIP_code': 'first',
        'Population': 'first',
        'Households': 'first',
        'Tram_Stations': 'first',
        'Primary_Roads': 'first',
        'Secondary_Roads': 'first',
        'Avg_Distance_km': 'mean',
        'H0': 'mean', 'H6': 'mean', 'H8': 'mean', 'H12': 'mean', 
        'H18': 'mean', 'H23': 'mean'  # Heures clés
    }).round(2)
    
    # Renommer les colonnes comme ton image
    df_master.columns = [
        'ZIP_code', 'Population', 'Households', 
        'Tram_Stations', 'Primary_Roads_km', 'Secondary_Roads_km',
        'Avg_Distance_km', 'H0_nuit', 'H6_matin', 'H8_pointe_matin',
        'H12_midi', 'H18_pointe_soir', 'H23_nuit'
    ]
    
    df_master.to_excel(writer, sheet_name='DONNEES_PAR_COMMUNE')
    
    # FEUILLE 2 : Trafic horaire complet (toutes les heures)
    print("→ Feuille 2: TRAFIC HORAIRE...")
    hourly_full = df_final.groupby('Commune')[[f'H{i}' for i in range(24)]].mean().round(2)
    hourly_full.to_excel(writer, sheet_name='TRAFIC_24H')
    
    # FEUILLE 3 : Par jour de la semaine
    print("→ Feuille 3: TRAFIC PAR JOUR...")
    daily_pivot = df_final.pivot_table(
        values='Avg_Distance_km', 
        index='Commune', 
        columns='Day', 
        aggfunc='mean'
    ).round(2)
    # Réordonner les jours
    day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    daily_pivot = daily_pivot[[d for d in day_order if d in daily_pivot.columns]]
    daily_pivot.to_excel(writer, sheet_name='TRAFIC_PAR_JOUR')
    
    # FEUILLE 4 : Priorités et recommandations
    print("→ Feuille 4: PRIORITÉS...")
    priority_data = df_master.copy()
    priority_data['Priorite'] = priority_data.apply(lambda x: 
        'HAUTE' if x['Avg_Distance_km'] > 7 and x['Tram_Stations'] == 0 else
        'MOYENNE' if x['Avg_Distance_km'] > 6 else
        'FAIBLE', axis=1
    )
    priority_data['Recommandation'] = priority_data.apply(lambda x:
        'Bus express urgent' if x['Priorite'] == 'HAUTE' else
        'Renforcement tram' if x['Tram_Stations'] > 0 and x['Avg_Distance_km'] > 7 else
        'Mobilité douce' if x['Avg_Distance_km'] < 5 else
        'Maintien', axis=1
    )
    priority_data.to_excel(writer, sheet_name='PRIORITES_ACTIONS')
    
    # FEUILLE 5 : Métadonnées (dictionnaire)
    print("→ Feuille 5: MÉTADONNÉES...")
    metadata = pd.DataFrame({
        'Colonne': df_master.columns.tolist() + [f'H{i}' for i in range(24)],
        'Description': [
            'Code postal de la commune',
            'Population totale (habitants)',
            'Nombre de ménages',
            'Nombre de stations de tramway',
            'Kilomètres de routes primaires',
            'Kilomètres de routes secondaires',
            'Distance moyenne de trajet (km)',
            'Trafic à minuit (km)', 'Trafic à 6h (km)', 'Trafic à 8h (km)',
            'Trafic à midi (km)', 'Trafic à 18h (km)', 'Trafic à 23h (km)'
        ] + [f'Distance moyenne à {i}h (km)' for i in range(24)],
        'Type': ['Numérique'] * (len(df_master.columns) + 24),
        'Source': ['Données statiques'] * 7 + ['Données GPS calculées'] * (6 + 24)
    })
    metadata.to_excel(writer, sheet_name='METADONNEES', index=False)

print(f"\n✅ FICHIER CRÉÉ: CASABLANCA_DATA_NETTOYEES.xlsx")
print("=" * 50)
print("📋 CONTENU:")
print("   1. DONNEES_PAR_COMMUNE → Synthèse par commune (comme ton image)")
print("   2. TRAFIC_24H → Toutes les heures H0-H23")
print("   3. TRAFIC_PAR_JOUR → Lundi au dimanche")
print("   4. PRIORITES_ACTIONS → Recommandations stratégiques")
print("   5. METADONNEES → Dictionnaire des variables")
print(f"\n🎉 Fichier prêt à l'emploi dans Excel!")
```

```python
print("=" * 70)
print("🔍 EXPLORATORY DATA ANALYSIS (EDA) - CASABLANCA")
print("=" * 70)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuration des graphiques
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# 1. CHARGEMENT DES DONNÉES
print("\n" + "=" * 70)
print("1️⃣ CHARGEMENT ET APERÇU DES DONNÉES")
print("=" * 70)

df = pd.read_excel('CASABLANCA_FINAL_PROPRE.xlsx', sheet_name='DONNEES_COMPLETES')
print(f"Dimensions: {df.shape[0]} communes × {df.shape[1]} variables")
print(f"\nColonnes: {list(df.columns)}")

# 2. STATISTIQUES DESCRIPTIVES
print("\n" + "=" * 70)
print("2️⃣ STATISTIQUES DESCRIPTIVES")
print("=" * 70)

# Sélectionner les colonnes numériques
numeric_cols = df.select_dtypes(include=[np.number]).columns
desc_stats = df[numeric_cols].describe().round(2)
print(desc_stats)

# 3. ANALYSE DES VALEURS MANQUANTES
print("\n" + "=" * 70)
print("3️⃣ ANALYSE DES VALEURS MANQUANTES")
print("=" * 70)

missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({
    'Valeurs manquantes': missing,
    'Pourcentage': missing_pct
})
print(missing_df[missing_df['Valeurs manquantes'] > 0] if missing.sum() > 0 else "✅ Aucune valeur manquante!")

# 4. DISTRIBUTION DES VARIABLES CLÉS
print("\n" + "=" * 70)
print("4️⃣ DISTRIBUTION DES VARIABLES CLÉS")
print("=" * 70)

fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 4.1 Distribution Population
axes[0,0].hist(df['Population'], bins=15, color='steelblue', edgecolor='black', alpha=0.7)
axes[0,0].set_title('Distribution de la Population', fontweight='bold')
axes[0,0].set_xlabel('Population')
axes[0,0].set_ylabel('Fréquence')

# 4.2 Distribution Distance moyenne
axes[0,1].hist(df['Avg_Distance_km'], bins=15, color='coral', edgecolor='black', alpha=0.7)
axes[0,1].set_title('Distribution Distance Moyenne', fontweight='bold')
axes[0,1].set_xlabel('Distance (km)')
axes[0,1].set_ylabel('Fréquence')

# 4.3 Boxplot Trafic par heure
trafic_data = df[['H0_nuit', 'H8_pointe_matin', 'H12_midi', 'H18_pointe_soir']]
axes[0,2].boxplot([df['H0_nuit'], df['H8_pointe_matin'], df['H12_midi'], df['H18_pointe_soir']], 
                  labels=['H0', 'H8', 'H12', 'H18'])
axes[0,2].set_title('Boxplot Trafic Horaire', fontweight='bold')
axes[0,2].set_ylabel('Distance (km)')

# 4.4 Superficie résidentielle
axes[1,0].hist(df['Residential area (m²)']/1e6, bins=15, color='green', edgecolor='black', alpha=0.7)
axes[1,0].set_title('Superficie Résidentielle', fontweight='bold')
axes[1,0].set_xlabel('Superficie (millions m²)')

# 4.5 Stations de tram
tram_counts = df['Tram_Stations'].value_counts().sort_index()
axes[1,1].bar(tram_counts.index, tram_counts.values, color='orange', edgecolor='black')
axes[1,1].set_title('Distribution Stations de Tram', fontweight='bold')
axes[1,1].set_xlabel('Nombre de stations')
axes[1,1].set_ylabel('Nombre de communes')

# 4.6 Routes primaires vs secondaires
axes[1,2].scatter(df['Primary_Roads_km'], df['Secondary_Roads_km'], 
                  s=100, alpha=0.6, c='purple', edgecolors='black')
axes[1,2].set_title('Routes Primaires vs Secondaires', fontweight='bold')
axes[1,2].set_xlabel('Routes primaires (km)')
axes[1,2].set_ylabel('Routes secondaires (km)')

plt.tight_layout()
plt.savefig('EDA_Distributions.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Graphiques sauvegardés: EDA_Distributions.png")

# 5. MATRICE DE CORRÉLATION
print("\n" + "=" * 70)
print("5️⃣ MATRICE DE CORRÉLATION")
print("=" * 70)

# Sélectionner variables pertinentes pour corrélation
corr_vars = ['Population', 'Residential area (m²)', 'Industrial area  (m²)', 
             'Tram_Stations', 'Primary_Roads_km', 'Secondary_Roads_km', 
             'Avg_Distance_km', 'H8_pointe_matin']
corr_matrix = df[corr_vars].corr().round(2)

plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='RdBu_r', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
            fmt='.2f', annot_kws={'size': 10})
plt.title('🔥 Matrice de Corrélation des Variables', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('EDA_Correlation.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Matrice de corrélation sauvegardée: EDA_Correlation.png")

# Corrélations significatives
print("\n📊 Corrélations significatives (|r| > 0.5):")
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        corr_val = corr_matrix.iloc[i, j]
        if abs(corr_val) > 0.5:
            print(f"  • {corr_vars[i]} ↔ {corr_vars[j]}: {corr_val:.2f}")

# 6. ANALYSE PAR CATÉGORIE
print("\n" + "=" * 70)
print("6️⃣ ANALYSE PAR CATÉGORIE (AVEC/SANS TRAM)")
print("=" * 70)

df['Avec_Tram'] = df['Tram_Stations'].apply(lambda x: 'Oui' if x > 0 else 'Non')

comparison = df.groupby('Avec_Tram').agg({
    'Population': ['mean', 'std'],
    'Avg_Distance_km': ['mean', 'std'],
    'H8_pointe_matin': ['mean', 'std'],
    'Residential area (m²)': 'mean'
}).round(2)

print(comparison)

# Test statistique (t-test)
avec_tram = df[df['Tram_Stations'] > 0]['Avg_Distance_km']
sans_tram = df[df['Tram_Stations'] == 0]['Avg_Distance_km']
t_stat, p_value = stats.ttest_ind(avec_tram, sans_tram)

print(f"\n📈 Test t de Student (distance moyenne):")
print(f"  • Avec tram: {avec_tram.mean():.2f} ± {avec_tram.std():.2f} km")
print(f"  • Sans tram: {sans_tram.mean():.2f} ± {sans_tram.std():.2f} km")
print(f"  • t-statistique: {t_stat:.3f}")
print(f"  • p-value: {p_value:.3f}")
print(f"  • Significatif: {'Oui' if p_value < 0.05 else 'Non'} (α = 0.05)")

# 7. DÉTECTION DES OUTLIERS
print("\n" + "=" * 70)
print("7️⃣ DÉTECTION DES OUTLIERS (MÉTHODE IQR)")
print("=" * 70)

def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

for col in ['Population', 'Avg_Distance_km', 'Residential area (m²)']:
    outliers, lb, ub = detect_outliers(df, col)
    print(f"\n{col}:")
    print(f"  • Bornes: [{lb:.0f}, {ub:.0f}]")
    print(f"  • Outliers détectés: {len(outliers)}")
    if len(outliers) > 0:
        print(f"  • Communes: {', '.join(outliers['Commune'].tolist())}")

# 8. PROFIL HORAIRE MOYEN
print("\n" + "=" * 70)
print("8️⃣ PROFIL HORAIRE MOYEN")
print("=" * 70)

heures = ['H0_nuit', 'H6_matin', 'H8_pointe_matin', 'H12_midi', 'H18_pointe_soir', 'H23_nuit']
profil_moyen = df[heures].mean()

plt.figure(figsize=(12, 6))
plt.plot(range(len(heures)), profil_moyen, marker='o', linewidth=3, markersize=10, color='red')
plt.fill_between(range(len(heures)), profil_moyen, alpha=0.3, color='red')
plt.xticks(range(len(heures)), ['00h', '06h', '08h', '12h', '18h', '23h'])
plt.xlabel('Heure de la journée', fontsize=12)
plt.ylabel('Distance moyenne (km)', fontsize=12)
plt.title('⏰ Profil Horaire Moyen - Toutes Communes', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
for i, v in enumerate(profil_moyen):
    plt.annotate(f'{v:.2f}', (i, v), textcoords="offset points", xytext=(0,10), ha='center')
plt.tight_layout()
plt.savefig('EDA_Profil_Horaire.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n📊 Statistiques du profil horaire:")
for h, val in zip(heures, profil_moyen):
    print(f"  • {h}: {val:.2f} km")

print("\n" + "=" * 70)
print("✅ EDA TERMINÉ - TOUS LES GRAPHIQUES SAUVEGARDÉS")
print("=" * 70)
print("Fichiers créés:")
print("  • EDA_Distributions.png")
print("  • EDA_Correlation.png")
print("  • EDA_Profil_Horaire.png")
```

```python
print("=" * 70)
print("🤖 MODÈLES PRÉDICTIFS - CASABLANCA")
print("=" * 70)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

# Chargement des données
df = pd.read_excel('CASABLANCA_FINAL_PROPRE.xlsx', sheet_name='DONNEES_COMPLETES')

# ============================================================
# MODÈLE 1: RÉGRESSION LINÉAIRE - Prédire la Distance Moyenne
# ============================================================
print("\n" + "=" * 70)
print("MODÈLE 1: RÉGRESSION LINÉAIRE MULTIPLE")
print("Objectif: Prédire Avg_Distance_km selon l'urbanisme")
print("=" * 70)

# Variables explicatives (X) et cible (y)
features = ['Population', 'Residential area (m²)', 'Industrial area  (m²)', 
            'Tram_Stations', 'Primary_Roads_km', 'Secondary_Roads_km']
X = df[features]
y = df['Avg_Distance_km']

# Division train/test (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalisation
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entraînement
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)

# Prédictions
y_pred_lr = lr_model.predict(X_test_scaled)

# Évaluation
r2_lr = r2_score(y_test, y_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
mae_lr = mean_absolute_error(y_test, y_pred_lr)

print(f"\n📊 Performance du modèle linéaire:")
print(f"  • R² (coefficient de détermination): {r2_lr:.3f}")
print(f"  • RMSE (erreur quadratique): {rmse_lr:.2f} km")
print(f"  • MAE (erreur absolue): {mae_lr:.2f} km")

# Coefficients (importance des variables)
coefficients = pd.DataFrame({
    'Variable': features,
    'Coefficient': lr_model.coef_,
    'Abs_Coef': np.abs(lr_model.coef_)
}).sort_values('Abs_Coef', ascending=False)

print(f"\n📈 Importance des variables (coefficients):")
for _, row in coefficients.iterrows():
    direction = "↑" if row['Coefficient'] > 0 else "↓"
    print(f"  • {row['Variable']}: {row['Coefficient']:+.3f} {direction}")

# ============================================================
# MODÈLE 2: RANDOM FOREST - Prédiction du Trafic de Pointe
# ============================================================
print("\n" + "=" * 70)
print("MODÈLE 2: RANDOM FOREST REGRESSOR")
print("Objectif: Prédire H8_pointe_matin (trafic matinal)")
print("=" * 70)

# Cible: trafic à 8h
y_rf = df['H8_pointe_matin']

X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(X, y_rf, test_size=0.3, random_state=42)

# Random Forest (pas besoin de normalisation)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Prédictions
y_pred_rf = rf_model.predict(X_test_rf)

# Évaluation
r2_rf = r2_score(y_test_rf, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test_rf, y_pred_rf))

print(f"\n📊 Performance Random Forest:")
print(f"  • R²: {r2_rf:.3f}")
print(f"  • RMSE: {rmse_rf:.2f} km")

# Importance des features
importance_rf = pd.DataFrame({
    'Variable': features,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print(f"\n🔥 Importance des variables (Random Forest):")
for _, row in importance_rf.iterrows():
    bar = "█" * int(row['Importance'] * 50)
    print(f"  • {row['Variable']:<25}: {row['Importance']:.3f} {bar}")

# ============================================================
# MODÈLE 3: CLUSTERING K-MEANS - Segmentation des Communes
# ============================================================
print("\n" + "=" * 70)
print("MODÈLE 3: K-MEANS CLUSTERING")
print("Objectif: Créer des profils de communes similaires")
print("=" * 70)

# Variables pour clustering
X_cluster = df[['Population', 'Avg_Distance_km', 'Tram_Stations', 'Residential area (m²)']].copy()
X_cluster_scaled = StandardScaler().fit_transform(X_cluster)

# Déterminer le nombre optimal de clusters (肘部法则/ silhouette)
from sklearn.metrics import silhouette_score
silhouettes = []
K_range = range(2, 6)
for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42)
    kmeans_temp.fit(X_cluster_scaled)
    silhouettes.append(silhouette_score(X_cluster_scaled, kmeans_temp.labels_))

optimal_k = K_range[np.argmax(silhouettes)]
print(f"\n📊 Nombre optimal de clusters: {optimal_k} (silhouette max: {max(silhouettes):.3f})")

# Clustering final
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_cluster_scaled)

# Analyse des clusters
print(f"\n🏷️ Caractéristiques des clusters:")
cluster_summary = df.groupby('Cluster').agg({
    'Commune': 'count',
    'Population': 'mean',
    'Avg_Distance_km': 'mean',
    'Tram_Stations': 'mean',
    'Residential area (m²)': 'mean'
}).round(2)
print(cluster_summary)

print(f"\n📋 Communes par cluster:")
for cluster in sorted(df['Cluster'].unique()):
    communes = df[df['Cluster'] == cluster]['Commune'].tolist()
    print(f"  Cluster {cluster}: {', '.join(communes)}")

# ============================================================
# VISUALISATIONS DES MODÈLES
# ============================================================
print("\n" + "=" * 70)
print("📊 VISUALISATIONS DES RÉSULTATS")
print("=" * 70)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Prédictions vs Réel (Régression Linéaire)
axes[0,0].scatter(y_test, y_pred_lr, alpha=0.7, s=100, c='blue', edgecolors='black')
axes[0,0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0,0].set_xlabel('Distance réelle (km)')
axes[0,0].set_ylabel('Distance prédite (km)')
axes[0,0].set_title(f'Régression Linéaire - Prédictions\nR² = {r2_lr:.3f}', fontweight='bold')
axes[0,0].grid(True, alpha=0.3)

# 2. Importance des variables (Random Forest)
axes[0,1].barh(importance_rf['Variable'], importance_rf['Importance'], color='green', alpha=0.7)
axes[0,1].set_xlabel('Importance')
axes[0,1].set_title('Importance des Variables\n(Random Forest)', fontweight='bold')
axes[0,1].grid(axis='x', alpha=0.3)

# 3. Clusters visualisation
scatter = axes[1,0].scatter(df['Population'], df['Avg_Distance_km'], 
                           c=df['Cluster'], s=200, alpha=0.7, cmap='viridis', edgecolors='black')
axes[1,0].set_xlabel('Population')
axes[1,0].set_ylabel('Distance moyenne (km)')
axes[1,0].set_title('Clusters de Communes\n(K-Means)', fontweight='bold')
plt.colorbar(scatter, ax=axes[1,0], label='Cluster')

# Ajouter noms des communes
for idx, row in df.iterrows():
    axes[1,0].annotate(row['Commune'], (row['Population'], row['Avg_Distance_km']),
                      xytext=(5, 5), textcoords='offset points', fontsize=7, alpha=0.7)

# 4. Comparaison des modèles
models = ['Régression\nLinéaire', 'Random\nForest']
r2_scores = [r2_lr, r2_rf]
axes[1,1].bar(models, r2_scores, color=['skyblue', 'lightgreen'], edgecolor='black', linewidth=2)
axes[1,1].set_ylabel('Score R²')
axes[1,1].set_title('Comparaison des Modèles\n(Performance)', fontweight='bold')
axes[1,1].set_ylim(0, 1)
for i, v in enumerate(r2_scores):
    axes[1,1].text(i, v + 0.02, f'{v:.3f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('Modeles_Predictions.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================
# PRÉDICTIONS POUR NOUVELLES COMMUNES (Exemple)
# ============================================================
print("\n" + "=" * 70)
print("🔮 EXEMPLE DE PRÉDICTION")
print("=" * 70)

# Simulation d'une nouvelle commune
nouvelle_commune = pd.DataFrame({
    'Population': [150000],
    'Residential area (m²)': [5000000],
    'Industrial area  (m²)': [500000],
    'Tram_Stations': [15],
    'Primary_Roads_km': [80],
    'Secondary_Roads_km': [60]
})

# Prédiction avec les deux modèles
nouvelle_scaled = scaler.transform(nouvelle_commune)
pred_distance = lr_model.predict(nouvelle_scaled)[0]
pred_trafic = rf_model.predict(nouvelle_commune)[0]

print(f"\n📍 Nouvelle commune simulée:")
print(f"  • Population: {nouvelle_commune['Population'][0]:,}")
print(f"  • Stations de tram: {nouvelle_commune['Tram_Stations'][0]}")
print(f"  • Superficie résidentielle: {nouvelle_commune['Residential area (m²)'][0]/1e6:.1f} M m²")

print(f"\n🔮 Prédictions:")
print(f"  • Distance moyenne estimée: {pred_distance:.2f} km")
print(f"  • Trafic matinal (8h) estimé: {pred_trafic:.2f} km")

# Classification du cluster
nouvelle_cluster = kmeans.predict(
    StandardScaler().fit_transform(nouvelle_commune[['Population', 'Avg_Distance_km', 
                                                     'Tram_Stations', 'Residential area (m²)']].fillna(0))
)[0]
print(f"  • Profil similaire aux communes: Cluster {nouvelle_cluster}")

print("\n" + "=" * 70)
print("✅ MODÉLISATION TERMINÉE!")
print("=" * 70)
print("Fichiers créés: Modeles_Predictions.png")
print("\n📊 Résumé des performances:")
print(f"  • Régression Linéaire: R² = {r2_lr:.3f} ({r2_lr*100:.1f}% de variance expliquée)")
print(f"  • Random Forest: R² = {r2_rf:.3f} ({r2_rf*100:.1f}% de variance expliquée)")
print(f"  • Clusters identifiés: {optimal_k} profils de communes")
```

```python
# ============================================================
# PRÉDICTIONS POUR NOUVELLES COMMUNES (Corrigé)
# ============================================================
print("\n" + "=" * 70)
print("🔮 EXEMPLE DE PRÉDICTION")
print("=" * 70)

# Simulation d'une nouvelle commune (données disponibles)
nouvelle_commune = pd.DataFrame({
    'Population': [150000],
    'Residential area (m²)': [5000000],
    'Industrial area  (m²)': [500000],
    'Tram_Stations': [15],
    'Primary_Roads_km': [80],
    'Secondary_Roads_km': [60]
})

# Prédiction avec les deux modèles
nouvelle_scaled = scaler.transform(nouvelle_commune)
pred_distance = lr_model.predict(nouvelle_scaled)[0]
pred_trafic = rf_model.predict(nouvelle_commune)[0]

print(f"\n📍 Nouvelle commune simulée:")
print(f"  • Population: {nouvelle_commune['Population'][0]:,}")
print(f"  • Stations de tram: {nouvelle_commune['Tram_Stations'][0]}")
print(f"  • Superficie résidentielle: {nouvelle_commune['Residential area (m²)'][0]/1e6:.1f} M m²")

print(f"\n🔮 Prédictions:")
print(f"  • Distance moyenne estimée: {pred_distance:.2f} km")
print(f"  • Trafic matinal (8h) estimé: {pred_trafic:.2f} km")

# Classification du cluster (utiliser seulement les variables disponibles + prédiction)
# On crée un profil avec la distance prédite pour trouver le cluster le plus proche
nouvelle_profile = np.array([[ 
    nouvelle_commune['Population'].iloc[0],
    pred_distance,  # On utilise la prédiction comme approximation
    nouvelle_commune['Tram_Stations'].iloc[0],
    nouvelle_commune['Residential area (m²)'].iloc[0]
]])

# Normaliser avec les mêmes paramètres que l'entraînement
scaler_cluster = StandardScaler()
scaler_cluster.fit(X_cluster)  # X_cluster contient les données d'entraînement
nouvelle_cluster_scaled = scaler_cluster.transform(nouvelle_profile)

# Prédire le cluster
nouvelle_cluster = kmeans.predict(nouvelle_cluster_scaled)[0]
print(f"  • Profil similaire aux communes: Cluster {nouvelle_cluster}")

# Afficher les communes similaires
communes_similaires = df[df['Cluster'] == nouvelle_cluster]['Commune'].tolist()
print(f"  • Communes comparables: {', '.join(communes_similaires)}")

print("\n" + "=" * 70)
print("✅ MODÉLISATION TERMINÉE!")
print("=" * 70)
```
