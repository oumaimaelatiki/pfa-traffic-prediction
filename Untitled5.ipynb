```python
import pandas as pd
import numpy as np

# Rechargement propre du fichier
df = pd.read_csv('Metro_Interstate_Traffic_Volume.csv', sep=';')

# Conversion date avec format européen (jour d'abord)
df['date_time'] = pd.to_datetime(df['date_time'], dayfirst=True)

# Gestion des jours fériés
df['holiday'] = df['holiday'].fillna('None')

# Correction des valeurs aberrantes
df.loc[df['temp'] < 200, 'temp'] = np.nan
df['temp'] = df['temp'].interpolate()
df.loc[df['rain_1h'] > 1000, 'rain_1h'] = 0

# Création des features temporelles
df['hour'] = df['date_time'].dt.hour
df['day_of_week'] = df['date_time'].dt.dayofweek
df['month'] = df['date_time'].dt.month
df['year'] = df['date_time'].dt.year
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
df['is_holiday'] = df['holiday'].apply(lambda x: 0 if x == 'None' else 1)

print("✅ Données nettoyées avec succès !")
print(f"Période couverte : {df['date_time'].min()} à {df['date_time'].max()}")
print(f"Nombre total de jours : {(df['date_time'].max() - df['date_time'].min()).days}")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration pour de belles figures (pour le rapport)
plt.style.use('seaborn-v0_8-darkgrid')  # Ou 'default' selon votre version
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Distribution avec courbe de densité
axes[0].hist(df['traffic_volume'], bins=50, color='steelblue', alpha=0.7, density=True)
df['traffic_volume'].plot(kind='density', ax=axes[0], color='red', linewidth=2)
axes[0].set_title('Distribution du volume de trafic', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Véhicules par heure')
axes[0].set_ylabel('Densité')
axes[0].grid(True, alpha=0.3)

# 2. Boxplot pour détecter les outliers
axes[1].boxplot(df['traffic_volume'], vert=True, patch_artist=True,
                boxprops=dict(facecolor='lightcoral', alpha=0.7))
axes[1].set_title('Boxplot du trafic\n(Détection des outliers)', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Véhicules par heure')
axes[1].grid(True, alpha=0.3)

# 3. Statistiques clés (texte sur le graphique)
stats_text = f"""
Statistiques descriptives :
• Moyenne : {df['traffic_volume'].mean():.0f} véh/h
• Médiane : {df['traffic_volume'].median():.0f} véh/h
• Écart-type : {df['traffic_volume'].std():.0f} véh/h
• Minimum : {df['traffic_volume'].min()} véh/h
• Maximum : {df['traffic_volume'].max()} véh/h
"""
axes[2].text(0.1, 0.5, stats_text, fontsize=12, family='monospace',
             verticalalignment='center', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
axes[2].axis('off')
axes[2].set_title('Résumé statistique', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('fig4_1_distribution.png', dpi=300, bbox_inches='tight')  # Pour le rapport
plt.show()

# Affichage des stats dans la console aussi
print("📊 ANALYSE UNIVARIÉE DU TRAFIC")
print("="*40)
print(f"Coefficient de variation : {(df['traffic_volume'].std()/df['traffic_volume'].mean())*100:.1f}%")
print(f"Skewness (asymétrie) : {df['traffic_volume'].skew():.2f}")
```

```python
# Calcul des moyennes par heure
hourly_stats = df.groupby('hour')['traffic_volume'].agg(['mean', 'std', 'min', 'max'])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Graphique 1 : Barres avec erreurs (écart-type)
bars = ax1.bar(hourly_stats.index, hourly_stats['mean'], 
               color=['darkred' if h in [7,8,9,16,17,18] else 'steelblue' for h in hourly_stats.index],
               alpha=0.8, edgecolor='black', linewidth=0.5)
ax1.errorbar(hourly_stats.index, hourly_stats['mean'], yerr=hourly_stats['std'], 
             fmt='none', color='black', alpha=0.5, capsize=3)
ax1.set_title('Volume moyen de trafic par heure\n(Heures de pointe en rouge)', 
              fontsize=14, fontweight='bold')
ax1.set_xlabel('Heure de la journée', fontsize=12)
ax1.set_ylabel('Véhicules par heure (moyenne)', fontsize=12)
ax1.set_xticks(range(0, 24))
ax1.grid(True, alpha=0.3, axis='y')

# Ajout des valeurs sur les barres de pointe
for hour in [7, 8, 9, 17, 18]:
    height = hourly_stats.loc[hour, 'mean']
    ax1.text(hour, height + 100, f'{height:.0f}', ha='center', va='bottom', fontweight='bold')

# Graphique 2 : Comparaison Semaine vs Weekend
weekday_hourly = df[df['is_weekend']==0].groupby('hour')['traffic_volume'].mean()
weekend_hourly = df[df['is_weekend']==1].groupby('hour')['traffic_volume'].mean()

ax2.plot(weekday_hourly.index, weekday_hourly.values, marker='o', linewidth=3, 
         label='Jours de semaine (Lun-Ven)', color='navy', markersize=6)
ax2.plot(weekend_hourly.index, weekend_hourly.values, marker='s', linewidth=3, 
         label='Weekend (Sam-Dim)', color='darkorange', markersize=6)
ax2.fill_between(weekday_hourly.index, weekday_hourly.values, alpha=0.2, color='navy')
ax2.fill_between(weekend_hourly.index, weekend_hourly.values, alpha=0.2, color='darkorange')

ax2.set_title('Comparaison des patterns horaires\nSemaine vs Weekend', 
              fontsize=14, fontweight='bold')
ax2.set_xlabel('Heure de la journée', fontsize=12)
ax2.set_ylabel('Véhicules par heure (moyenne)', fontsize=12)
ax2.legend(loc='upper left', fontsize=10)
ax2.grid(True, alpha=0.3)
ax2.set_xticks(range(0, 24))

plt.tight_layout()
plt.savefig('fig4_2_rush_hours.png', dpi=300, bbox_inches='tight')
plt.show()

# Calcul des stats pour le rapport
morning_rush = df[df['hour'].isin([7,8,9])]['traffic_volume'].mean()
evening_rush = df[df['hour'].isin([16,17,18])]['traffic_volume'].mean()
night_low = df[df['hour'].isin([2,3,4])]['traffic_volume'].mean()
weekend_reduction = ((df[df['is_weekend']==0]['traffic_volume'].mean() - 
                     df[df['is_weekend']==1]['traffic_volume'].mean()) / 
                     df[df['is_weekend']==0]['traffic_volume'].mean() * 100)

print("\n📈 RÉSULTATS CLÉS POUR LE RAPPORT :")
print("="*50)
print(f"Heures de pointe matinales (7h-9h) : {morning_rush:.0f} véh/h en moyenne")
print(f"Heures de pointe soir (16h-18h) : {evening_rush:.0f} véh/h en moyenne")
print(f"Trafic nocturne (2h-4h) : {night_low:.0f} véh/h en moyenne")
print(f"Réduction weekend vs semaine : {weekend_reduction:.1f}%")
```

```python
# Code corrigé et complet pour la matrice de corrélation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sélection des variables
corr_vars = ['traffic_volume', 'hour', 'day_of_week', 'month', 'temp', 
             'rain_1h', 'snow_1h', 'clouds_all', 'is_weekend', 'is_holiday']

# Calcul de la matrice
correlation_matrix = df[corr_vars].corr()

# Création de la figure avec meilleure configuration
plt.figure(figsize=(14, 12))

# Heatmap COMPLÈTE (sans masque pour voir tout)
sns.heatmap(correlation_matrix, 
            annot=True,           # Valeurs affichées
            fmt='.2f',            # 2 décimales
            cmap='RdBu_r',        # Rouge-Bleu inversé
            center=0,             # Centre sur 0
            square=True,          # Cases carrées
            linewidths=0.5,       # Lignes fines
            annot_kws={"size": 10}, # Taille des nombres
            cbar_kws={"shrink": 0.8, "label": "Corrélation"})

plt.title('Matrice de corrélation des variables\n(Focus sur traffic_volume)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('fig4_4_correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# Tableau des corrélations avec traffic_volume (TRIÉ)
print("\n" + "="*60)
print("📊 CORRÉLATIONS AVEC TRAFFIC_VOLUME (triées par force)")
print("="*60)
correlations = correlation_matrix['traffic_volume'].drop('traffic_volume').sort_values(key=abs, ascending=False)

for var, corr in correlations.items():
    force = "🔴 FORTE" if abs(corr) > 0.5 else "🟡 MODÉRÉE" if abs(corr) > 0.2 else "⚪ FAIBLE"
    direction = "positive" if corr > 0 else "négative"
    print(f"{var:15s} : {corr:+.3f}  ({force} {direction})")
```

```python
# VERSION CORRIGÉE - Matrice de corrélation complète
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Variables à analyser
corr_vars = ['traffic_volume', 'hour', 'day_of_week', 'month', 'temp', 
             'rain_1h', 'snow_1h', 'clouds_all', 'is_weekend', 'is_holiday']

# Calcul
corr_matrix = df[corr_vars].corr()

# FIGURE 1: Heatmap complète avec masque triangle supérieur
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, 
            mask=mask,
            annot=True, 
            fmt='.2f',
            cmap='RdBu_r',
            center=0,
            vmin=-1, vmax=1,
            square=True,
            linewidths=0.5,
            cbar_kws={"shrink": 0.8},
            annot_kws={'size': 11})

plt.title('Matrice de Corrélation des Variables', fontsize=14, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('fig4_4_correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# FIGURE 2: Barplot des corrélations avec traffic_volume (PLUS LISIBLE)
plt.figure(figsize=(10, 6))
corr_with_target = corr_matrix['traffic_volume'].drop('traffic_volume').sort_values()

colors = ['darkred' if x < -0.2 else 'red' if x < 0 else 'lightblue' if x < 0.2 else 'darkblue' 
          for x in corr_with_target]

bars = plt.barh(corr_with_target.index, corr_with_target.values, color=colors, edgecolor='black')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.xlabel('Coefficient de corrélation avec traffic_volume', fontsize=12)
plt.title('Corrélations avec le volume de trafic (variable cible)', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)

# Ajout des valeurs sur les barres
for bar, val in zip(bars, corr_with_target.values):
    plt.text(val + (0.01 if val >= 0 else -0.01), bar.get_y() + bar.get_height()/2, 
             f'{val:.2f}', va='center', ha='left' if val >= 0 else 'right', fontweight='bold')

plt.tight_layout()
plt.savefig('fig4_4_correlations_barplot.png', dpi=300, bbox_inches='tight')
plt.show()

# Affichage numérique clair
print("\n" + "="*55)
print("📊 CORRÉLATIONS AVEC TRAFFIC_VOLUME")
print("="*55)
for var, val in corr_with_target.sort_values(ascending=False).items():
    print(f"{var:15s} : {val:+.3f}")
```
